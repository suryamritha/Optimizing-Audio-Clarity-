```python
import os
import numpy as np
import librosa
import librosa.display
import IPython.display as ipd
import matplotlib.pyplot as plt

# Function to load data and labels
def load_data(directory):
    data = []
    labels = []
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith('.wav'):
                file_path = os.path.join(root, file)
                label = os.path.basename(root)
                data.append(file_path)
                labels.append(label)
    return data, labels

# Function for spectral subtraction noise reduction with automatic noise spectrum identification
def spectral_subtraction(segmented_audio, sr, top_db=20):
    stft = librosa.stft(segmented_audio)
    magnitude = np.abs(stft)
    
    # Identify noise profile by finding segments with lowest energy
    noise_threshold = np.percentile(magnitude, 20)  # Using 20th percentile as noise threshold
    noise_frames = magnitude[:, magnitude.mean(axis=0) < noise_threshold]
    noise_spectrum = np.mean(noise_frames, axis=1) if noise_frames.size > 0 else np.zeros(magnitude.shape[0])
    
    clean_spectrum = np.maximum(magnitude - noise_spectrum[:, np.newaxis], 0)
    cleaned_audio = librosa.istft(clean_spectrum * np.exp(1j * np.angle(stft)))
    return cleaned_audio

# Function to perform voice activity detection (VAD) and segment the audio using spectral subtraction
def perform_vad_and_segmentation(audio, sr, vad_threshold=20, min_segment_duration=0.1):
    non_silent_intervals = librosa.effects.split(audio, top_db=vad_threshold)
    segmented_audio = []
    for start, end in non_silent_intervals:
        segment_duration = (end - start) / sr
        if segment_duration >= min_segment_duration:
            segment = audio[start:end]
            cleaned_segment = spectral_subtraction(segment, sr)
            segmented_audio.append(cleaned_segment)
    return segmented_audio

# Function to plot waveform and spectrogram
def plot_waveform_and_spectrogram(audio, sr, title):
    plt.figure(figsize=(12, 8))
    plt.subplot(2, 1, 1)
    plt.plot(np.arange(len(audio)) / sr, audio)
    plt.title(title + ' Waveform')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')
    plt.subplot(2, 1, 2)
    spec = librosa.amplitude_to_db(np.abs(librosa.stft(audio)), ref=np.max)
    librosa.display.specshow(spec, sr=sr, x_axis='time', y_axis='log')
    plt.title(title + ' Spectrogram')
    plt.colorbar(format='%+2.0f dB')
    plt.tight_layout()
    plt.show()

# Function to compute metrics of silence in audio
def compute_silence_metrics(audio, sr, rms_threshold=0.01, amplitude_threshold_db=-60, zcr_threshold=0.1, spectral_contrast_threshold=25):
    rms_energy = librosa.feature.rms(y=audio)
    amplitude_db = librosa.amplitude_to_db(np.abs(audio), ref=np.max)
    zcr = librosa.feature.zero_crossing_rate(audio)
    spectral_contrast = librosa.feature.spectral_contrast(y=audio, sr=sr)
    stft = np.abs(librosa.stft(audio))
    spectral_centroid = np.mean(librosa.feature.spectral_centroid(S=stft, sr=sr))
    spectral_flatness = np.mean(librosa.feature.spectral_flatness(y=audio))
    spectral_bandwidth = np.mean(librosa.feature.spectral_bandwidth(y=audio, sr=sr))
    chroma_feature = np.mean(librosa.feature.chroma_stft(y=audio, sr=sr), axis=1)
    tonnetz = np.mean(librosa.feature.tonnetz(y=audio, sr=sr), axis=1)
    tempo, _ = librosa.beat.beat_track(y=audio, sr=sr)
    
    rms_silence_frames = np.where(rms_energy < rms_threshold)[1]
    amplitude_silence_frames = np.where(amplitude_db < amplitude_threshold_db)[0]
    zcr_silence_frames = np.where(zcr < zcr_threshold)[1]
    spectral_contrast_silence_frames = np.where(np.mean(spectral_contrast, axis=0) < spectral_contrast_threshold)[0]
    
    silence_frames = np.union1d(rms_silence_frames, amplitude_silence_frames)
    silence_frames = np.union1d(silence_frames, zcr_silence_frames)
    silence_frames = np.union1d(silence_frames, spectral_contrast_silence_frames)
    
    total_duration = librosa.get_duration(y=audio, sr=sr)
    silence_duration = len(silence_frames) / float(sr)
    silence_percentage = (silence_duration / total_duration) * 100

    return {
        'silence_duration': silence_duration,
        'silence_percentage': silence_percentage,
        'spectral_centroid': spectral_centroid,
        'rms_energy': np.mean(rms_energy),
        'amplitude_db': np.mean(amplitude_db),
        'zcr': np.mean(zcr),
        'spectral_contrast': np.mean(spectral_contrast),
        'spectral_flatness': spectral_flatness,
        'spectral_bandwidth': spectral_bandwidth,
        'chroma_feature': chroma_feature.tolist(),  # Convert to list for better printing
        'tonnetz': tonnetz.tolist(),  # Convert to list for better printing
        'tempo': tempo
    }

# Main function
if __name__ == "__main__":
    # Load audio data
    file_path = r"C:\Users\surya\Downloads\rain.wav"
    y, sr = librosa.load(file_path, sr=None)
    print("Playing original audio...")
    ipd.display(ipd.Audio(y, rate=sr))

    # Perform voice activity detection and segmentation on the original audio using spectral subtraction
    segmented_audio = perform_vad_and_segmentation(y, sr)

    # Compute silence metrics for each segment
    silence_metrics = []
    for i, segment in enumerate(segmented_audio):
        print(f"Computing silence metrics for segment {i + 1}...")
        metrics = compute_silence_metrics(segment, sr)
        silence_metrics.append(metrics)
        print(f"Segment {i + 1} Metrics:", metrics)

    # Combine all segmented audio into a single array
    combined_audio = np.concatenate(segmented_audio)

    # Compute silence metrics for combined audio
    combined_metrics = compute_silence_metrics(combined_audio, sr)
    print("\nCombined Audio Metrics:")
    for key, value in combined_metrics.items():
        print(f"{key}: {value if not isinstance(value, list) else value}")

    # Compute silence metrics for original audio
    original_metrics = compute_silence_metrics(y, sr)
    print("\nOriginal Audio Metrics:")
    for key, value in original_metrics.items():
        print(f"{key}: {value if not isinstance(value, list) else value}")

    # Play the combined audio
    print("Playing combined audio...")
    ipd.display(ipd.Audio(combined_audio, rate=sr))

    # Plot waveform and spectrogram of the original and combined audio
    plot_waveform_and_spectrogram(y, sr, "Original Audio")
    plot_waveform_and_spectrogram(combined_audio, sr, "Combined Audio")

```
